## **6. Model 1: Linear Regression**

We used K-fold cross validation method here. Instead of testing the model only once, we split the data into 5 equal parts (folds). We train the model 5 times, and each time, a different part is used for testing. This ensures our accuracy score (R**2) is stable and not just a "lucky" result from a single split.

1. lr_model = LinearRegression() 📏 --> Creates a new instance of the Linear Regression model
2. kf = KFold(n_splits=5, shuffle=True, random_state=42) 🔄 --> Sets up the 5-fold cross-validation by shuffling the data and splitting it into 5 groups.
3. cv_scores = cross_val_score(lr_model, X, y, cv=kf, scoring='r2') 🧪 --> Runs the model 5 times on different parts of the data and calculates the R2score for each run.
4. np.mean(cv_scores)🎯 --> Calculates the Average Accuracy across all folds.
5. np.std(cv_scores)📉 --> Calculates the Standard Deviation to measure stability.


The Mean (np.mean): Represents the overall performance. In the project, the mean is 0.8827, meaning the model explains about 88% of the sales on average.

The Standard Deviation (np.std): Represents the Reliability. A low value (like 0.0395 in this data) means the model is very stable and performs consistently across different parts of the dataset.

A high mean and a low standard deviation prove that the Linear Regression model is both accurate and robust. It isn't just "lucky" with one part of the data; it works well across the whole set.

*Key Differences: Linear Regression vs. Random Forest 🔍*

Feature ----------- Linear Regression (📏) ----------------------- Random Forest (🌲)

## **10. Save the Model (The "Pickle" Step)**

Once we are happy with our models, we need to save them. Think of this as "exporting" the intelligence we've built. We use the joblib library to convert our Python objects into .pkl files. This allows us to load the models instantly into a web app or another script without retraining them from scratch.

1. import joblib 📦 --> Imports the toolkit for serializing and saving large Python objects.
2. joblib.dump(lr_model, ..)💾 --> Writes the Linear Regression model to a permanent file.
3. joblib.dump(rf_model, ...) 💾 --> Writes the Random Forest model to a permanent file.
4. print("Success! ...") 🎉 --> Confirms that the files have been successfully created.

Why this step is important?🎯


*Portability*: You can now move these .pkl files to a server or a Streamlit app.

*Efficiency*: Loading a saved model takes milliseconds, while training one might take much longer with larger datasets.

*Deployment*: This is the bridge between a "Data Science Notebook" and a "Real Product."
